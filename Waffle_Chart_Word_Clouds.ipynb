## Waffle Chart

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
fig = plt.figure(
    figsize=(8,6),
    FigureClass = Waffle,
    rows = 3,
    values = country_dic.Num,
    labels = list(country_dic.Country)
)
```

## Advantages:
* More visually attractive.
* Used for attractive Dashboards.

## Disadvantages:
* Waffle Chart becomes too complicated when too many segments are involved.
* individualized spaces do not leave a spot to put numbers or text within the visual itself.

```python
#a = ['a','b', 'c', 'a', 'd', 'c' ,'c']
dic = {}
for w in words:
    if w not in dic:
        dic[w] = 1
    else:
        dic[w] = dic[w] + 1
    print(dic)
```

```python
swd = sorted(dic.items(), key = lambda e: e[1], reverse=True)
```

```python
swd[0:10]
```

```python
with open('jieba_dictionary/stopwords.txt','r',encoding='utf-8') as file:
    stopwords = file.read().split('\n')
print('停用詞:'+'/'.join(stopwords))
```

```python
dic = {}
for w in words:
    if (w not in stopwords) and len(w) > 2:
        if w not in dic:
            dic[w] = 1
        else:
            dic[w] = dic[w] + 1
```

```python
swd = sorted(dic.items(), key = lambda e: e[1], reverse=True)
```

```python
swd[0:10]
```

```python
word_cloud = ''
for i in swd:
    word_cloud = word_cloud+i[0]+' '
```

```python
word_cloud
```

```python
type(word_cloud)
```

```python
cloud = WordCloud().generate(word_cloud)
```

```python
#cloud2 = WordCloud().generate(text1)  # 使用 generate() function 就可以將文字自動統計、製作文字雲。
```

```python
import matplotlib.pyplot as plt
plt.figure(
    figsize=(8,6))
plt.imshow(cloud)
```
