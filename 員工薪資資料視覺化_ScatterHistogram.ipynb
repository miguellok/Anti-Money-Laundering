# Histogram (直方圖)：如果要統計連續性的變數(x軸)，可以使用直方圖，它會自動分組距(bins)，它只能針對一個變數作統計。

```python
plt.figure(figsize=(8,6))
my_bins = 10
arr = plt.hist(df['公司年度資訊-平均員工薪資費用(仟元/人)']/10,bins=my_bins,range=(50,300))
for i in range(my_bins):
    plt.text(arr[1][i],arr[0][i],'     '+str(int(arr[0][i])))
plt.ylabel('公司數',fontsize=15)
plt.xlabel('平均員工薪資費用(萬)',fontsize=15)
plt.yticks(np.arange(0, 20,2))
plt.grid()
```
